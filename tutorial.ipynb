### Import relevent modules

```python
import pawlikMorphLSST as pmlsst
import matplotlib.pyplot as plt
from astropy.io import fits
%matplotlib inline
```

### Load image

```python
img = fits.getdata("sample/sdsscutout_211.51-0.31_rband.fits")
plt.imshow(img)
```

```python
imgnew, mask = pmlsst.prepareimage(img)
fig, axs = plt.subplots(1, 2)
axs[0].imshow(mask)
axs[1].imshow(imgnew)
```

```python
params = pmlsst.calculateMorphology(imgnew, mask)
```

```python
print(f"A = {params[0][0]}")
print(f"Abgr = {params[0][1]}")
print(f"As = {params[1][0]}")
print(f"As90 = {params[2][0]}")
```
